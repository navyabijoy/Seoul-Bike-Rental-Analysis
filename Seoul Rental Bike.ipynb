```python
from sklearn.model_selection import train_test_split
```

```python
df['Date'] = pd.to_datetime(df['Date'])
```

```python
df.sort_values(by='Date', inplace=True)
```

```python
df
```

```python
df['Date'].value_counts()
```

```python
split_date = '2018-06-01'  # Choose a date that makes sense for your data

train_data = df[df['Date'] < split_date]
test_data = df[df['Date'] >= split_date]

# Print the sizes of the training and testing sets
print("Training set size:", len(train_data))
print("Testing set size:", len(test_data))
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error
#import matplotlib.pyplot as plt
```

```python
df.columns
```

```python
features = ['Hour', 'Temperature(°C)', 'Humidity(%)', 'Wind speed (m/s)', 'Visibility (10m)', 
            'Dew point temperature(°C)', 'Solar Radiation (MJ/m2)', 'Rainfall(mm)', 'Snowfall (cm)', 
            'Seasons', 'Holiday', 'Functioning Day', 'Day', 'Month', 'Year']

# Target variable
target = 'Rented Bike Count'

# Training set
X_train = train_data[features]
y_train = train_data[target]

# Testing set
X_test = test_data[features]
y_test = test_data[target]

# Initialize and train the model
model = LinearRegression()
model.fit(X_train, y_train)

# Make predictions on the testing set
y_pred = model.predict(X_test)

# Evaluate the model
mse = mean_squared_error(y_test, y_pred)
print(f'Mean Squared Error (MSE): {mse}')

# Plot actual vs predicted values
plt.figure(figsize=(12, 6))
plt.plot(test_data['Date'], y_test, label='Actual')
plt.plot(test_data['Date'], y_pred, label='Predicted')
plt.title('Actual vs Predicted Rented Bike Count')
plt.xlabel('Date')
plt.ylabel('Rented Bike Count')
plt.legend()
plt.show()
```

```python
season_mapping = {'Spring': 1, 'Summer': 2, 'Autumn': 3, 'Winter': 4}

df['Seasons'] = df['Seasons'].map(season_mapping)
```

```python
features = ['Hour', 'Temperature(°C)', 'Humidity(%)', 'Wind speed (m/s)', 'Visibility (10m)', 
            'Dew point temperature(°C)', 'Solar Radiation (MJ/m2)', 'Rainfall(mm)', 'Snowfall (cm)', 
            'Seasons', 'Holiday', 'Functioning Day', 'Day', 'Month', 'Year']

# Target variable
target = 'Rented Bike Count'

# Split the data into training and testing sets
train_data, test_data = train_test_split(df, test_size=0.2, shuffle=False)  # Adjust the split ratio and shuffle as needed

# Training set
X_train = train_data[features]
y_train = train_data[target]

# Testing set
X_test = test_data[features]
y_test = test_data[target]

# Initialize and train the model
model = LinearRegression()
model.fit(X_train, y_train)

# Make predictions on the testing set
y_pred = model.predict(X_test)

# Evaluate the model
mse = mean_squared_error(y_test, y_pred)
print(f'Mean Squared Error (MSE): {mse}')

# Plot actual vs predicted values
plt.figure(figsize=(12, 6))
plt.plot(test_data['Date'], y_test, label='Actual')
plt.plot(test_data['Date'], y_pred, label='Predicted')
plt.title('Actual vs Predicted Rented Bike Count')
plt.xlabel('Date')
plt.ylabel('Rented Bike Count')
plt.legend()
plt.show()
```

```python
# Assuming your DataFrame is named df
# Assuming the DataFrame is already split into training and testing sets

# Feature selection
features = ['Hour', 'Wind speed (m/s)', 'Visibility (10m)', 
            'Dew point temperature(°C)', 'Solar Radiation (MJ/m2)', 'Rainfall(mm)', 'Snowfall (cm)', 
            'Seasons', 'Holiday', 'Functioning Day', 'Day', 'Month', 'Year']

# Target variable
target = 'Rented Bike Count'

# Training set
X_train = train_data[features]
y_train = train_data[target]

# Testing set
X_test = test_data[features]
y_test = test_data[target]
```

```python
# Initialize and train the Random Forest model
model = RandomForestRegressor(n_estimators=100, random_state=42)
model.fit(X_train, y_train)
```

```python
# Make predictions on the testing set
y_pred = model.predict(X_test)
```

```python
# Evaluate the model using Mean Squared Error (MSE)
mse = mean_squared_error(y_test, y_pred)

print(f'Mean Squared Error (MSE): {mse}')
```
